```python
import pandas as pd
import tensorflow as tf
from tensorflow import keras
target=pd.read_csv('target.csv')
features=pd.read_csv('features.csv')
```

```python
features=pd.get_dummies(features)
features
```

```python
test=pd.read_csv('test.csv')
test.head()
```

```python
from sklearn.multiclass import OneVsRestClassifier
from sklearn.svm import SVC 
```

```python
clf = OneVsRestClassifier(SVC()).fit(features, target)
```

```python
predit=clf.predict(test)
```

```python
print(predit[:20])
```

```python
from sklearn.neural_network import MLPClassifier
clf = MLPClassifier(random_state=1, max_iter=2000000).fit(features, target)
```

```python
predMLP=clf.predict(test)
```

```python
predMLP[:10]
```

```python
prdMLP = pd.DataFrame(predMLP, columns = ['pred_ldlhigh',	'pred_ldllow',	'ldlok',	'hgbhigh',	'hgblow',	'hgbok',	'hdlhigh',	'hdllow',	'hdlok'])
```
