```python
import pandas as pd
import numpy as np
```

```python
train_data=pd.read_csv('Train_trimmed.csv')
```

```python
train_data.head()
```

```python
train_data=train_data.dropna()
train_data=train_data.drop(columns=['Reading_ID'])
train_data.head()
```

```python
from sklearn import preprocessing
le = preprocessing.LabelEncoder()
train_data['hdl']=le.fit_transform(train_data['hdl_cholesterol_human'])
train_data['hgb']=le.fit_transform(train_data['hemoglobin(hgb)_human'])
train_data['ldl']=le.fit_transform(train_data['cholesterol_ldl_human'])
```

```python
train_data.head()
```

```python
train_data=train_data.drop(columns=['hdl_cholesterol_human','hemoglobin(hgb)_human','cholesterol_ldl_human'])
```

```python
train_data.head()
```

```python
features=train_data.drop(columns=['hdl','hgb','ldl'])
```

```python
features.head()
```

```python
train_data.head()
```

```python
target = train_data[{"ldl","hgb","hdl"}]
```

```python
target.head()
```

```python
target.shape
```

```python
features.shape
```

```python
from sklearn.cluster import KMeans
```

```python
kmeans = KMeans(n_clusters=3, random_state=0).fit(features)
```

```python
kmean_label=kmeans.labels_
```

```python
print(kmean_label[:100])
```

```python
kmean_label = pd.DataFrame(kmean_label, columns = ['Kmean_cluster'])
```

```python
kmean_label.size
```

```python
features['Kmean_cluster']=kmean_label
```

```python
features.head()
```

```python
from sklearn.preprocessing import LabelBinarizer
y=target['hdl']
y_dense = LabelBinarizer().fit_transform(y)
```

```python
hgb = pd.DataFrame(y_dense, columns = ['hdlhigh','hdllow','hdlok'])
```

```python
hgb
```

```python
target["hdlhigh"]=hgb["hdlhigh"]
target["hdllow"]=hgb["hdllow"]
target["hdlok"]=hgb["hdlok"]
```

```python
target.head()
```

```python
target=target.drop(columns=['hgb','ldl','hdl'])
```

```python
target.head()
```

```python
features.head()
```

```python
target.to_csv('target.csv', index=False)
```
